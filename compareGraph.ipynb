```python
import scipy.io as sio
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
gf = pd.read_csv('Dataset/socfb-CMU.csv')
```

```python
def read_graph(filename):
    # read graph from edgelist file
    g = nx.Graph()
    
    f = open(filename)
    for l in f:
        l = l.rstrip().replace(" ", ";").replace(",", ";").replace("\t", ";").split(";")
        g.add_edge(l[0], l[1])
    return g
G = read_graph('socfb-CMU.csv')
```

```python
def average_degree(filename):
    N,K = G.order(), G.size()
    return float(K) / N
degree = average_degree('socfb-CMU.csv')
degree
```

```python
G_random = nx.gnp_random_graph(6621,0.0111)
G_barabasi = nx.barabasi_albert_graph(6621, 38)
```

```python
def degree_distribution(g,g2):
    # get the degree histogram
    hist = nx.degree_histogram(g)
    hist_1 = nx.degree_histogram(g2)
    
    plt.plot(range(0, len(hist)), hist, ".", color='blue')
    plt.plot(range(0, len(hist_1)), hist_1, ".", color='orange')
    plt.title("Degree Distribution")
    plt.xlabel("Degree")
    plt.ylabel("#Nodes")
    plt.loglog()
    plt.show()
```

```python
degree_distribution(G,G_random)
```

```python
degree_distribution(G,G_barabasi)
```
